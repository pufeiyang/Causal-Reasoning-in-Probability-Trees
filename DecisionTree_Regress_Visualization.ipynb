# 导入必要库

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import pandas as pd
import numpy as np
from sklearn import tree
```

# 导入数据集

```python
path = '/DataSetClean.xlsx'
df = pd.read_excel(path)
df.head(5)
```

# 预处理数据

```python
del df['class']
df.head(5)
```

```python
features_name = df.iloc[:,:-3].columns.values.tolist()

features = df.iloc[:,:-3].values
target1 = df.iloc[:,-3].values.reshape(-1,1)
print('The shape of features:',features.shape)
print('The shape of MAP(max):',target1.shape)

features[features=='女'] = 0
features[features=='男'] = 1
features[features=='是'] = 1
features[features=='否'] = 0
features[features=='I'] = 1
features[features=='II'] = 2
features[features=='Ⅱ'] = 2
features[features=='III'] = 3
features[features=='Ⅲ'] = 3
```

```python
min_max_scaler = preprocessing.MinMaxScaler()

std_features = min_max_scaler.fit_transform(features)
std_target1 = min_max_scaler.fit_transform(target1)

# ori_data = min_max_scaler.inverse_transform(std_data) 
```

# 训练模型

```python
X_train, X_test, Y_train, Y_test = train_test_split(std_features, std_target1, 
                              train_size=0.8, random_state=0)

print('The shape of X_train:',X_train.shape)
print('The shape of X_test:',X_test.shape)
print('The shape of Y_train:',Y_train.shape)
print('The shape of Y_test:',Y_test.shape)
```

```python
model = DecisionTreeRegressor(random_state = 0)

model.fit(X_train, Y_train)

Y_pre = model.predict(X_test)
```
